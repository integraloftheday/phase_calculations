## Gibbs Free Energy of Ice and Water
Atmospheric Pressure 

```python
T = np.linspace(-10, 10, 100) + 273.15
P = 0.1 * np.ones(1)

PT = np.array([P, T], dtype=object)  # Removed dtype=object
out_ice1 = sf.getProp(PT, "Ih")  # Ensure 'II' is a valid phase
out_lwater = sf.getProp(PT,"water1")
```

```python
plt.plot(T-273.15,out_ice1.G[0])
plt.plot(T-273.15,out_lwater.G[0])
plt.legend(["Ice 1", "Liquid Water"])
plt.xlabel("Temperature C")
plt.ylabel("Gibs Free Energy J/kg")
```
